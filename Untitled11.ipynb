```python
#W6A1
def removing(input_string):
  seen = set()
  numbers = [int(x) for x in input_string.split()]
  result = []
  for num in numbers :
    if num not in seen :
      seen.add(num)
      result.append(num)
  return result
  print (result)
if __name__ == "__main__":
    input_data = input("Nhập một dòng gồm các số nguyên (cách nhau bởi khoảng trắng): ")

    print(removing(input_data))


```

```python
#W6A2
def calculate_cumulative_sum(input_string):
    try:
        numbers = [int(x) for x in input_string.split()]
    except ValueError:
        print("Lỗi: Đầu vào chứa ký tự không phải là số nguyên.")
        return []

    result = []
    current_sum = 0

    for num in numbers:
        current_sum += num
        result.append(current_sum)
    return result

if __name__ == "__main__":

    input_data = input("Nhập một dòng gồm các số nguyên (cách nhau bởi khoảng trắng): ")
    final_result = calculate_cumulative_sum(input_data)
    print(final_result)
```

```python
#W6A9
d1 = dict(item.split(":") for item in input("Nhập dict 1: ").split())
d2 = dict(item.split(":") for item in input("Nhập dict 2: ").split())

d1 = {k: int(v) for k, v in d1.items()}
d2 = {k: int(v) for k, v in d2.items()}

result = {}

for k, v in d1.items():
    result[k] = v

for k, v in d2.items():
    if k in result:
        result[k] += v
    else:
        result[k] = v

result = dict(sorted(result.items()))
print("Dictionary sau khi gộp:", result)
```

```python
#W6A10
nums = list(map(int, input("Nhập list số nguyên: ").split()))
k = int(input("Nhập k: "))
pairs = []
for i in range(len(nums)):
    for j in range(i + 1, len(nums)):
        if nums[i] + nums[j] == k:
            pairs.append((nums[i], nums[j]))
pairs.sort(key=lambda x: x[0])
print(pairs)
```

```python
#W6A18
n, m, k = map(int, input().split())

a = []
for i in range(n):
    row = list(map(int, input().split()))
    a.append(row)

# Tổng cột k
sum_col = 0
for i in range(n):
    sum_col += a[i][k]

# Tổng hàng k
sum_row = 0
for j in range(m):
    sum_row += a[k][j]

print("Tổng cột k:", sum_col)
print("Tổng hàng k:", sum_row)
```

```python
#W6A19
def vi_tri_max(a):
    max_val = a[0]
    pos = 0

    for i in range(1, len(a)):
        if a[i] > max_val:
            max_val = a[i]
            pos = i

    return pos
a = list(map(int, input("Nhập dãy số nguyên: ").split()))
pos = vi_tri_max(a)
print("Vị trí của phần tử lớn nhất là:", pos)
print("Giá trị lớn nhất là:", a[pos])
```

```python
#W6A20
def tim_vi_tri_k(a, k):
    for i in range(len(a)):
        if a[i] == k:
            return i
    return -1
a = list(map(int, input("Nhập dãy số nguyên: ").split()))
k = int(input("Nhập số k: "))
print(tim_vi_tri_k(a, k))
```

```python
#W6A21
s = input("Nhập các chuỗi: ").split()

result = {}

for i in range(len(s)):
    word = s[i]
    if word not in result:
        result[word] = []
    result[word].append(i)

for word in result:
    result[word] = tuple(result[word])

print(result)
```
