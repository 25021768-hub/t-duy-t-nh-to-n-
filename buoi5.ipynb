```python
#W5A1
def Max_Of_Two(a: int, b: int) -> int:
    if a > b:
        return a
    elif b > a:
        return b
    else:
        return a
print(Max_Of_Two(2,5))
```

```python
#W5A3
import math
def is_prime(x):
  limit = int(math.sqrt(x)) + 1
  for i in range(2, limit):
    if x % i == 0 :
      return False
    return True
print(is_prime(11))
```

```python
#W5A4
import math

def is_perfect(n: int) -> bool:
    if n <= 0:
        raise ValueError("Số cần kiểm tra phải là số nguyên dương (n > 0).")

    if n == 1:
        return False
    sum_of_divisors = 1
    limit = int(math.sqrt(n))

    for i in range(2, limit + 1):
        if n % i == 0:
            sum_of_divisors += i
            divisor_pair = n // i
            if divisor_pair != i:
                sum_of_divisors += divisor_pair
    return sum_of_divisors == n

a = int(input())
print(is_perfect(a))
```

```python
#W5A5
def ktra(arr, k):
    if k not in arr:
        return -1
    else:
        return arr.index(k) + 1
a = [int(x) for x in input().split()]
k = int(input())
print(ktra(a, k))
```

```python
#W5A6
def factorial(n: int) -> int:
    if n < 0:
        raise ValueError("Giai thừa chỉ được tính cho các số nguyên không âm (n >= 0).")
    if n == 0 or n == 1:
        return 1
    result = 1
    for i in range(2, n + 1):
        result *= i

    return result
a = int(input())
print(factorial(a))
```

```python
#W5A7
import math
def simply_computer(num1 : float, num2 : float, operat : str):
  result = None
  if operat == '+':
    result = num1 + num2
  elif operat == '-':
    result = num1 - num2
  elif operat == '*':
    result = num1 * num2
  elif operat == '/':
    if num2 != 0:
      result = num1 / num2
    else:
      print("Không thể chia cho 0")
  elif operat == '^':
    result = num1 ** num2
  if result is not None:
    formatted = "{:.2f}".format(result)
    return formatted

  return None

a = float(input())
ope = input()
b = float(input())
print(simply_computer(a, b, ope))

```

```python
#W5A8
def hamming(n1,n2):
  bin_n1 = bin(n1)[2:]
  bin_n2 = bin(n2)[2:]

  len_n1 = len(bin_n1)
  len_n2 = len(bin_n2)
  max_len = max(len_n1, len_n2)

  bin_n1 = bin_n1.zfill(max_len)
  bin_n2 = bin_n2.zfill(max_len)

  distance = 0
  for i in range(max_len):
    if bin_n1[i] != bin_n2[i]:
      distance += 1

  return distance

a = int(input())
b = int(input())
print(hamming(a,b))
```

```python
#W5A9
def sum_digit_math(n):
  if n < 0 :
    n = abs(n)
  total = 0
  while n > 0 :
    digit = n % 10
    total += digit
    n //= 10

  return total
a = int(input())
print(sum_digit_math(a))
```
